```python
import plotly.express as px 
import plotly.graph_objects as go
from plotly.data import gapminder
```

```python
country_stats = gapminder()
```

```python
country_stats.describe()
```

```python
country_stats
```

```python
country_stats["country"].unique()
```

```python
px.line(country_stats, x="year", y="lifeExp", color="country")
```

```python
population_per_country = country_stats.groupby(["year", "continent"])["pop"].sum().reset_index()
population_per_country
fig = px.bar(country_stats, x="continent", y="pop", color="country", animation_frame="year", range_y=[0, 3_500_000_000])
fig
```

```python
fig = px.density_heatmap(country_stats, x="lifeExp", y="gdpPercap")
fig
```

```python
fig = px.line(country_stats, x="year", y="lifeExp", color="country", facet_row="continent")
fig
```
